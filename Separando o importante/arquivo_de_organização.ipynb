## Talvez, importante, a reconstrução de imagem de uma forma geral.

```python
import numpy as np
import matplotlib.pyplot as plt

# Gerar uma imagem de exemplo (simulando uma imagem original)
image_size = 256  # Tamanho da imagem (256x256 pixels)
image = np.zeros((image_size, image_size))  # Inicializa uma imagem vazia
image[96:160, 96:160] = 1  # Adiciona um quadrado branco como exemplo

# Visualizar a imagem original
plt.figure()
plt.title("Imagem Original")
plt.imshow(image, cmap="gray")
plt.colorbar()
plt.show()

# Passo 1: Gerar o espaço k (domínio de Fourier)
k_space = np.fft.fft2(image)  # Transformada de Fourier 2D
k_space = np.fft.fftshift(k_space)  # Desloca o zero para o centro

# Visualizar o espaço k (magnitude)
plt.figure()
plt.title("Espaço k (Fourier)")
plt.imshow(np.log(np.abs(k_space) + 1), cmap="gray")  # Escala log para visualizar melhor
plt.colorbar()
plt.show()

# Passo 2: Reconstruir a imagem do espaço k
reconstructed_image = np.fft.ifft2(np.fft.ifftshift(k_space))  # Transformada de Fourier inversa
reconstructed_image = np.abs(reconstructed_image)  # Magnitude para a reconstrução

# Visualizar a imagem reconstruída
plt.figure()
plt.title("Imagem Reconstruída")
plt.imshow(reconstructed_image, cmap="gray")
plt.colorbar()
plt.show()
```
